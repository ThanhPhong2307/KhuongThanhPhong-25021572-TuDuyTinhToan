W3A16

```python
def lam_tron_xuong(x: float) -> int:
    x_int = int(x)
    if x < 0 and x != x_int:
        return x_int - 1
    else:
        return x_int

def lam_tron_len(x: float) -> int:
    x_int = int(x)
    if x > 0 and x != x_int:
        return x_int + 1
    else:
        return x_int

def lam_tron_gan(x: float) -> int:
    if x >= 0:
        return int(x + 0.5)
    else:
        return int(abs(x) + 0.5) * -1
def lam_tron():
    try:
        so_thuc = float(input("Nhập vào một số thực: "))
    except ValueError:
        print("Lỗi: Đầu vào không phải là số thực hợp lệ. Vui lòng thử lại.")
        return
    ket_qua_len = lam_tron_len(so_thuc)
    ket_qua_xuong = lam_tron_xuong(so_thuc)
    ket_qua_gan = lam_tron_gan(so_thuc)
    print(f"\n--- KẾT QUẢ ---")
    print(f"Số thực đã nhập: {so_thuc}")
    print(f"Làm tròn lên (Ceiling):   {ket_qua_len}")
    print(f"Làm tròn xuống (Floor):  {ket_qua_xuong}")
    print(f"Làm tròn gần nhất (Round): {ket_qua_gan}")
if __name__ == "__main__":
    lam_tron()
```
