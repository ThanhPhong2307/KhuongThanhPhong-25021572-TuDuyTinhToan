```python
def so_chan(x):
    for i in range(x + 1):
        if i % 2 == 0:
            print(i, end=' ')
so_chan(100)
```

```python
def so_duong_chan(x):
    for i in range(x + 1):
        if i % 3 == 0 and i % 2 == 0:
            print(i, end=' ')
so_duong_chan(100)
```

```python
a , b = map(int, input().split())
uoc_chung = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        uoc_chung = i - 1 #Loại 0 ra (vì 0 không được tính là ước chung)
print(uoc_chung)
```

Bài 14

```python
max = float('-inf')
while True:
    try:
     n = int(input())
    except ValueError:
        continue
    if n == -1:
        break
    if n > max:
        max = n
print(f"{max}")
    
```
