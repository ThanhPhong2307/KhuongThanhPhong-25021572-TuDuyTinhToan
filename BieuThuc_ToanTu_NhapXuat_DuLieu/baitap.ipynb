#Bài 6

```python
import math
a, b, c = map(int, input().split())
#áp dụng công thức Heron tính diện tích tam giác
p = (a + b + c) / 2
dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
#áp dụng định lý tổng 2 cạnh trong tam giác luôn lớn hơn cạnh còn lại
if a + b < c:
    print("Khong phai 3 canh cua tam giac")
elif a + c < b:
    print("Khong phai 3 canh cua tam giac")
elif b + c < a:
    print("Khong phai 3 canh cua tam giac")
else:
    print(f"Là 3 cạnh của tam giác.\nDiện tích của tam giác là {dien_tich:.1f}")
```
