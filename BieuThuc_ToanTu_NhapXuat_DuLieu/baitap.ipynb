```python
a,b = map(int, input().split())
#Diện tích trồng cây = diện tích hình vuông (ab) - diện tích hình tròn (3.14r^2)
#ta có chiều rộng a, chiều dài b -> r = 0.5a
dien_tich = a * b - ((0.5 * a) ** 2) * 3.14
print(f"diện tích trồng cây xung quanh là:{dien_tich:.2f}.")
```

#Bài 6

```python
import math
a, b, c = map(int, input().split())
p = (a + b + c) / 2
dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
if a + b < c:
    print("Khong phai 3 canh cua tam giac")
elif a + c < b:
    print("Khong phai 3 canh cua tam giac")
elif b + c < a:
    print("Khong phai 3 canh cua tam giac")
else:
    print(f"Là 3 cạnh của tam giác.\nDiện tích của tam giác là {dien_tich:.1f}")
```
